**Deskripsi Kasus**

Banyak orang di Indonesia telah terinfeksi COVID-19 sejak awal tahun 2020. Wilayah sekitar DKI Jakarta seperti Jawa Barat dan Banten juga mencatat banyak kasus, menunjukkan bahwa orang banyak bergerak di sana, yang mempercepat penyebaran virus.

**Memvisualisasikan data menggunakan diagram Plot**

```python
# Plot Kasus COVID-19 yang dikonfirmasi per provinsi
plt.figure(figsize=(10,6))
sns.barplot(y='province_name', x='confirmed', data=df)
plt.title('Jumlah Kasus COVID-19 Terkonfirmasi per Provinsi')
plt.xlabel('Kasus Terkonfirmasi')
plt.ylabel('Provinsi')
plt.show()
```

```python
# Hubungan antara populasi dan kasus yang dikonfirmasi
plt.figure(figsize=(10,6))
sns.scatterplot(x='population', y='confirmed', data=df)
plt.title('Hubungan antara Populasi dan Kasus Terkonfirmasi')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Contoh data
data = {
    'province_name': ['DKI Jakarta', 'Jawa Barat', 'Banten', 'Jawa Timur', 'Jawa Tengah'],
    'confirmed': [598, 98, 84, 66, 43],
    'deceased': [51, 14, 4, 4, 6],
    'released': [31, 5, 1, 8, 0]
}
df = pd.DataFrame(data)

# Membuat plot batang untuk masing-masing variabel
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(18, 6))

# Plot kasus konfirmasi
df.plot(x='province_name', y='confirmed', kind='bar', ax=axes[0], color='skyblue', legend=False)
axes[0].set_title('Kasus Konfirmasi per Provinsi')
axes[0].set_xlabel('Nama Provinsi')
axes[0].set_ylabel('Jumlah')
axes[0].tick_params(axis='x', rotation=45)

# Plot kasus kematian
df.plot(x='province_name', y='deceased', kind='bar', ax=axes[1], color='salmon', legend=False)
axes[1].set_title('Kasus Kematian per Provinsi')
axes[1].set_xlabel('Nama Provinsi')
axes[1].set_ylabel('jumlah')
axes[1].tick_params(axis='x', rotation=45)

# Plot kasus pemulihan
df.plot(x='province_name', y='released', kind='bar', ax=axes[2], color='lightgreen', legend=False)
axes[2].set_title('Kasus Pemulihan per Provinsi')
axes[2].set_xlabel('nama provinsi')
axes[2].set_ylabel('Jumlah')
axes[2].tick_params(axis='x', rotation=45)

# Menyesuaikan tampilan plot
plt.tight_layout()
plt.show()
```

```python
# Pie chart proporsi provinsi dengan jumlah kasus tertinggi
top_provinces = df.nlargest(5, 'confirmed')
plt.figure(figsize=(8,8))
plt.pie(top_provinces['confirmed'], labels=top_provinces['province_name'], autopct='%1.1f%%', startangle=140)
plt.title('5 Provinsi dengan Kasus Terkonfirmasi Covid-19 Tertinggi')
plt.show()
```
